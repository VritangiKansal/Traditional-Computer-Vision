```python
# necessary imports

import numpy as np
import matplotlib.pyplot as plt
import skimage
import math
import scipy
import cv2 as cv
from PIL import Image
```

<B>Q1. Perform Gaussian smoothing on the grayscale image affleck_gray.png (or
affleck_gray_flip.png). Try with multiple sigma values, starting with larger values
(e.g., 20 to .5). When does the face become recognizable to your friends?<B>

```python
# read the image
faceIm = plt.imread('affleck_gray.png')

# display the image and set the colormap as gray
plt.imshow(faceIm, cmap='gray')
plt.axis('off')
plt.show()
```

```python
''' plot the output of gaussian filter applied on original image
    with sigma values changing from 0.5 to 20. '''

fig = plt.figure(figsize=(20, 20))
rows = 7
col = 7
count = 1
#for sigma in [float(j) / 2 for j in range(40, -1, -1)]:
for sigma in range(40, -1, -2):
    
    mask_size = math.ceil(3*sigma)*2 + 1
    
    # returns Gaussian Filter coefficients 
    u = cv.getGaussianKernel(mask_size, sigma)
    # create the kernel
    kernel = np.outer(u, u)
    # convolve gaussian kernel wirh the image using mode as nearest which is same as replicate in matlab
    gauss_2D_convolve = scipy.ndimage.convolve(faceIm, kernel, mode='nearest')
    
    if(sigma == 22):
        image_to_save = gauss_2D_convolve

    fig.add_subplot(rows, col, count)
    count+=1
    
    title_string = "Sigma: " + str(sigma)
    plt.title(title_string)
    plt.imshow(gauss_2D_convolve, cmap='gray')
```

```python
# write the image which is formed by sigma value of 22
image_to_save = cv.normalize(image_to_save, None, 0, 255, cv.NORM_MINMAX, cv.CV_8U)
Image.fromarray(image_to_save).save('gIm.bmp')
```

- Gaussian Smoothing has been applied to the image with sigma values starting from 40 to going down at 0. 
- I showed this set of 21 images, starting from sigma 40, to a friend and asked him to guess the actor. It seemed absurd at first and he fumbled, but eventually guessed the actor at sigma = 22. 
- We can see from these images how with gradual increase in kernel size, the detail and background babble in the images is getting removed. 
- Bigger the sigma, bigger the kernel size and hence it takes more time for code to run. This particular set of jupyter cell takes like ~10 minutes to run since it has to evaluate all the gaussian kernels and then convolve plus fit in the subplots. 



<B>2) Write a function to compute and display the 2D Gaussian derivative masks Gx and
Gy for a given sigma. Sample the Gaussian derivative/gradient (2D) equation
directly (see class notes) at the appropriate x,y (col, row!) locations. Note: each mask
is a square 2D matrix. Please ensure that the positive derivative lobe is on the side of
the increasing direction of each axis for each mask (see notes regarding the negative
sign in the equations). Plot each mask (either in 2D or 3D). [3 pts]<B>

```python
''' function to compute the 2D Gaussian Derivate mask

    converted the matlab version into python present in 
    
    https://github.com/stsreyas/cvLib/blob/master/gaussDeriv2D.m '''


def gaussDeriv2D(sigma):
    maskSize = math.ceil(3*sigma)*2 + 1
    
    # The mask is generated from -3*sigma to 3*sigma
    startIdx = -1*math.ceil(3*sigma);
    endIdx = math.ceil(3*sigma);
    
    # The center location is 0, which we can gather from the above start and end indices
    centerLoc = 0
    
    
    # The factor A for generating the masks
    A = 1/((2*math.pi)*(sigma**2))
    
    
    # Creating square matrices G, Gx and Gy with size maskSize
    G = np.zeros((maskSize, maskSize))
    Gx = np.zeros((maskSize, maskSize))
    Gy = np.zeros((maskSize, maskSize))
        
    # Iterating from -3*sigma to 3*sigma. Note: here we consider r as y and c as x in our formula

    y = -3*sigma;
    sigmaSq = sigma**2

    for r in range(0, maskSize):
        x = -3*sigma
        for c in range(0, maskSize):
            # Using the formula described at the top we can generate values of
            # Gx and Gy
            distx = (x - centerLoc)
            disty = (y - centerLoc)
            GExp = (math.exp(-1*(distx**2)/(2*sigmaSq))) * (math.exp(-1*(disty**2)/(2*sigmaSq)))
            G[r, c] = A * GExp
            Gx[r, c] = ((distx)/(2*sigmaSq**2))*GExp
            Gy[r, c] = ((disty)/(2*sigmaSq**2))*GExp
            x = x + 1
        y = y + 1
        
    return Gx, Gy
```

```python
sigma = 5
Gx, Gy = gaussDeriv2D(sigma)

''' plot the masks in 2D'''

fig = plt.figure(figsize=(10,20))
rows = 1
cols = 2
fig.add_subplot(rows,cols,1)
plt.title("Gx")
plt.imshow(Gx, cmap='gray')
fig.add_subplot(rows,cols,2)
plt.title("Gy")
plt.imshow(Gy, cmap='gray')
```

- Gaussian Mask has been calculated for sigma = 5
- Because of the high value of sigma, thicker gradient ridges should appear on the images. 
- The negative sign has been removed from the equation of Gx[r, c] and Gy[r, c] so that the positive derivative lobe is on the side of the increasing direction of each axis. 
- 2D plots of the masks have been drawn. 

<B>3) Compute and display the gradient magnitude of an image - search the web for an
interesting image that has strong vertical and horizontal boundaries/edges; convert to
grayscale if necessary (you know how to do this!); make sure to upload the image
with your code in the Carmen submission. [2 pts] </B>

```python
# upload an image 
updowIm = plt.imread('upside_down.jpeg')

# convert rgb to grayscale
updow_gray_Im = skimage.color.rgb2gray(updowIm)

# display the image with grayscale colormap
plt.imshow(updow_gray_Im, cmap='gray')
plt.show()
```

```python
# convolve the gaussian matrices over x and y axis over the image  
gxIm = scipy.ndimage.convolve(updow_gray_Im, Gx, mode='nearest')
gyIm = scipy.ndimage.convolve(updow_gray_Im, Gy, mode='nearest')
magIm = np.sqrt(np.multiply(gxIm, gxIm) + np.multiply(gyIm, gyIm))

# plot the figures
fig = plt.figure(figsize=(10,10))
rows = 2
cols = 2
fig.add_subplot(rows,cols,1)
plt.title("gxIm")
plt.imshow(gxIm, cmap='gray')
fig.add_subplot(rows,cols,2)
plt.title("gyIm")
plt.imshow(gyIm, cmap='gray')
fig.add_subplot(rows,cols,3)
plt.title("magIm")
plt.imshow(magIm, cmap='gray')
fig.add_subplot(rows,cols,4)
plt.title("Original")
plt.imshow(updow_gray_Im, cmap='gray')
```

- The selected image is infamously called as [topsy-turvy-truck](https://mymodernmet.com/rick-sullivan-upside-down-truck/).
- gxIm captures the changes in the horizontal direction of the image. See how hardly nothing is captured inside the body of the image here since not much is changing in the x direction. Though it captures the wheel outline well since there's difference in gradient value there. 
- gyIm captures the change in the vertical direction of the image. There's a change in the magnitude of pixels as we move from top to bottom within the body of the vehicle and the similar has been captured here as well. The image has definitely more changes in the vertical direction and hence gyIm captures somewhat the better outline of the original image as compared to gxIm.
- magIm cobines the results of gxIm and gyIm. It produces thicker gradient ridges due to high sigma value used in the mask. 

<B>4) Threshold and display the “gradient magnitude” image with different threshold T
levels. [2 pts] <B>

```python
''' look at the grayscale histogram of the image and try to identify what grayscale ranges corresponds to
    different edges in the image or background '''

histogram, bin_edges = np.histogram(magIm, bins=256, range=(0.0, 1))

fig, ax = plt.subplots()
plt.plot(bin_edges[0:-1], histogram)
plt.title("Grayscale Histogram")
plt.xlabel("grayscale value")
plt.ylabel("pixels")
plt.xlim(0, 0.2)
```

```python
fig = plt.figure(figsize=(10,10))
rows = 2
cols = 2

t1 = np.median(magIm)
t2 = np.mean(magIm)
t3 = 0.05
t4 = 0.075

tIm1 = magIm > t1
tIm2 = magIm > t2
tIm3 = magIm > t3
tIm4 = magIm > t4

fig.add_subplot(rows,cols,1)
plt.title("Thresholding with median value")
plt.imshow(tIm1, cmap='gray')
fig.add_subplot(rows,cols,2)
plt.title("Thresholding with mean value")
plt.imshow(tIm2, cmap='gray')
fig.add_subplot(rows,cols,3)
plt.title("Thresholding with t = 0.05")
plt.imshow(tIm3, cmap='gray')
fig.add_subplot(rows,cols,4)
plt.title("Thresholding with t = 0.075")
plt.imshow(tIm4, cmap='gray')
```

- Grayscale histogram is drawn to see where the pixel values peek and get an idea for thresholding values. 
- For thresholding, four values have been chosen:
    - median of all the pixel values: this could capture the car well but def has lot of noise too
    - mean of all the pixel values: captures less noise than median but still captures some extra edges
    - 0.05: from graph it looked like the fair half: gave a good result in terms of what's visible in the image
    - 0.075: only very prominent edges are visible

<B>5) Compare the above results with the Sobel masks. [2 pts]<B>

```python
# finds the vertical edges of an image
fxIm = skimage.filters.sobel_v(updow_gray_Im)
# finds the horizontal edges of an image
fyIm = skimage.filters.sobel_h(updow_gray_Im)
magIm = np.sqrt(np.multiply(fxIm, fxIm) + np.multiply(fyIm, fyIm))
T = 0.25
tIm = magIm > T

# plot the figures
fig = plt.figure(figsize=(10,10))
rows = 2
cols = 2
fig.add_subplot(rows,cols,1)
plt.title("fxIm")
plt.imshow(fxIm, cmap='gray')
fig.add_subplot(rows,cols,2)
plt.title("fyIm")
plt.imshow(fyIm, cmap='gray')
fig.add_subplot(rows,cols,3)
plt.title("magIm")
plt.imshow(magIm, cmap='gray')
fig.add_subplot(rows,cols,4)
plt.title("Thresholding with t = 0.25")
plt.imshow(tIm, cmap='gray')
```

- Firstly, I felt that the Sobel filter was less sensitive to background noise even while producing fxIm and fyIm. 
- The above hypothesis might be true since Sobel filter includes part smoothing while trying to detect edges. 
- Sobel was also successful in making edges more visible maybe because it has provided more weights to it's immediate neighbours.

<B>6.Run the MATLAB canny edge detector, edge(Im,'canny'), on your image and display
the default results. How does it compare? (Python: you can use the scikit-image
package) [2 pts]<B>

```python
edges = skimage.feature.canny(updow_gray_Im, sigma = 1, low_threshold=0.25, high_threshold=0.8)
plt.imshow(edges, cmap='gray')
plt.show()
```

- In Sobel the edges are rough. Whereas, the Canny produces smoother edges due to the implementation of Non-maxima suppression and thresholding.
- I feel like Canny gave better detailing if you see the outline inside wheels and the front of the vehicle. 
- If I would have applied blur on the image then extra background noise would have been removed from the output image and Canny would have given even better results.
