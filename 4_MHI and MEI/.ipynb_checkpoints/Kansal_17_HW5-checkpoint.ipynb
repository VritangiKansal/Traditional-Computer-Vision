Q1. Using the images (aerobic-[001-022].bmp) provided on the class materials site,
experiment with simple “motion detection” between consecutive frames using (abs)
image differencing. Clean-up and remove any tiny regions (e.g., use techniques such
as bwareaopen, median filtering, etc.). Experiment with different thresholds. [2 pts]

```python
fig = plt.figure(figsize=(20, 20))
rows = 6
col = 6
num = 1

threshold = 0.1

img_array = []

image_1 = plt.imread('aerobic-001.bmp')
image_1 = skimage.img_as_float(image_1)

for count in range(2, 23):
    
    image_name = 'aerobic-0' + str(count).zfill(2) + '.bmp'
    image_2 = plt.imread(image_name)
    image_2 = skimage.img_as_float(image_2)
    
    abs_img_difference = np.abs(image_2 - image_1)
    
    img_closing = skimage.morphology.closing(abs_img_difference, skimage.morphology.square(3))
    
    img_thresh = img_closing > threshold
    
    img_array.append(img_thresh)
    
    fig.add_subplot(rows, col, num)
    
    title_string = "img" + str(num) + " - img" + str(num + 1) 
    plt.title(title_string)
    plt.imshow(img_thresh, cmap='gray')
    
    num+=1
    
    image_1 = image_2
    
img_array = np.array(img_array)
```

- Consecutive image differences have been plotted. 
- Threshold value of 0.1 is taken in the above visualisation after trying other thresholds. 
- Closing operation is performed on the threshold images so that we have one huge component to detect motion. 

```python
# experiment with thresh levels 

threshold_ranges = [0.02, 0.05, 0.1, 0.15, 0.2, 0.25, 0.3, 0.5]

image_2 = plt.imread('aerobic-021.bmp')
image_2 = skimage.img_as_float(image_2)


abs_img_difference = np.abs(image_2 - image_1)
    
img_closing = skimage.morphology.closing(abs_img_difference,skimage.morphology.square(3))

fig = plt.figure(figsize=(15, 15))
rows = 2
col = 4
num = 1

for i in threshold_ranges:
    img_thresh = img_closing > i
    
    fig.add_subplot(rows, col, num)
    
    title_string = "threshold:" + str(i)
    plt.title(title_string)
    plt.imshow(img_thresh, cmap='gray')
    
    num+=1
    

```

- Multiple threshold levels were experimented on the difference image.
- We can see that thresholds like 0.02 and 0.05 have lot of noise in it whereas threshold level above 0.25 leads to missing of the required data.

```python
# compute and plot MHI

MHI_array = np.zeros(img_array.shape)


fig = plt.figure(figsize=(20, 20))
rows = 6
col = 6
num = 1

for i in range(0, 21):
    for x in range(0, 320):
        for y in range(0, 240):
            if img_array[i][x][y] == 1:
                MHI_array[i][x][y] = 22
            else:
                MHI_array[i][x][y] = max(0, MHI_array[i-1][x][y] - 1)
    fig.add_subplot(rows, col, num)
    title_string = "img" + str(i+1+1)
    plt.title(title_string)
    plt.imshow(MHI_array[i], cmap='gray')
    num+=1
```

MHI produces images through which we can capture better movement. See how the faded area transformation indicates the direction of motion here. 

```python
# simultitude moments

def cal_moments(image, i, j, mean_x, mean_y, x, y):
    
    return np.sum((pow((x - mean_x), i)) * (pow((y - mean_y), j)) * image) / pow(np.sum(image), ((i+j)/ 2) + 1)

def simultitude_moment(image):
    
    y,x = np.mgrid[:image.shape[0],:image.shape[1]]
   
    
    moments = []
    
    mean_x = np.sum(x*image)/np.sum(image)
    mean_y = np.sum(y*image)/np.sum(image)
    
        
    # find moments
    for i in range(0, 4):
        for j in range(0, 4):
            if i+j == 2 or i+j ==3:
                value = cal_moments(image, i, j, mean_x, mean_y, x, y)
                moments.append(value)
    return moments
```

```python
# normalize data

def normalize_data(data):
    normal = np.zeros(data.shape)
    for i in range(data.shape[0]):
        for j in range(data.shape[1]):
            normal[i][j] = max(0, data[i][j]*20.0-1.0)/21.0
    return normal
```

```python
# silmultitude moments of the last MEI

MEI_normalized = normalize_data(MEI_array[20])
simultitude_moment(MEI_normalized)
```

```python
# silmultitude moments of the last MHI

MHI_normalized = normalize_data(MHI_array[20])
simultitude_moment(MHI_normalized)
```

Q3. Create a 101x101 image with a black (0) background and a white (255) box of size
21x21, placing the upper-left corner at pixel (row=40, col=6). Create another new box
image, but shift the box 1-pixel to the right and 1-pixel down. Compute the normal
flow between the images. Use 3x3 Sobel Fx, Fy gradient masks (with the proper
normalization values – see notes) and a 3x3 average mask on each image when
computing Ft. Use MATLAB’s quiver function to draw the motion vectors on the
image (call imagesc, then ‘hold on’, and lastly call quiver). (Make sure your gradient
mask orientations/directions [and the plot axes] are consistent and point in the proper
directions!!!) Is the result what you expected? Why or why not? Comment on the
flow for the 4 sides of the box and also for the 4 corners. [5 pts]

```python
# create two boxes

box1 = np.zeros((101,101))
box1[39:60, 5:26] = 255
box1 = skimage.img_as_float(box1)

box2 = np.zeros((101,101))
box2[40:61, 6:27] = 255
box2 = skimage.img_as_float(box2)

f = plt.figure()
f.add_subplot(1,2, 1)
plt.imshow(box1, cmap='gray')
f.add_subplot(1,2, 2)
plt.imshow(box2, cmap='gray')
```

```python
# find normal optic flow between them

from scipy import signal
x_mask = 0.25 * np.array([[-1., 1.], [-1., 1.]])
y_mask = 0.25 * np.array([[-1., -1.], [1., 1.]])
t_mask = 0.25 * np.array([[1., 1.], [1., 1.]])
x_mask = np.fliplr(x_mask)
    
mode = 'same'
fx = (signal.convolve2d(box1, x_mask, boundary='symm', mode=mode))
fy = (signal.convolve2d(box1, y_mask, boundary='symm', mode=mode))
ft = (signal.convolve2d(box2, t_mask, boundary='symm', mode=mode) + signal.convolve2d(box1, -t_mask, boundary='symm', mode=mode))
    
u = np.zeros(box1.shape)
v = np.zeros(box1.shape)
    
window = np.ones((3,3))
    
denom = (signal.convolve2d(fx**2, window, boundary='symm', mode=mode) * signal.convolve2d(fy**2, window, boundary='symm', mode=mode) -signal.convolve2d(fx*fy, window, boundary='symm', mode=mode)**2)
denom[denom == 0] = 1
    
u = ((signal.convolve2d(fy**2, window, boundary='symm', mode=mode) * signal.convolve2d(fy*ft, window, boundary='symm', mode=mode) + signal.convolve2d(fx*fy, window, boundary='symm', mode=mode) * signal.convolve2d(fy*ft, window, boundary='symm', mode=mode) ) /denom)
v = ((signal.convolve2d(fx*ft, window, boundary='symm', mode=mode) * signal.convolve2d(fx*fy, window, boundary='symm', mode=mode) -signal.convolve2d(fx**2, window, boundary='symm', mode=mode) * signal.convolve2d(fy*ft, window, boundary='symm', mode=mode)) / denom)
    
```

```python
x = np.arange(0, box1.shape[1], 1)
y = np.arange(0, box1.shape[0], 1)
x, y = np.meshgrid(x, y)
p = 3

plt.figure(figsize=(10,10))
plt.imshow(box1, cmap='gray')
plt.quiver(x[::p, ::p], y[::p, ::p],u[::p, ::p], v[::p, ::p],
color = 'red', pivot = 'middle', headwidth=3, headlength=3, scale=30)
```

- See how vectors on the boundary are pointing in the horizontal direction.
- 3 vectors in the corner have both x-y components. 
- Bottom right corner vector has only y component in the normal optic flow.
